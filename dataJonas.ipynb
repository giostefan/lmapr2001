```python
import numpy as np
import matplotlib.pyplot as plt
with open("dataJonas.txt") as file : 
    kappa = np.zeros(4000)
    omega = np.zeros(4000)
    n = np.zeros(4000)
    lines = file.readlines()
    for i in range(7999) :
        if i%2 == 0 : 
            data = lines[i].strip().split()
            j = i//2
            kappa[j] = data[0]
            omega[j] = data[1]
            n[j] = data[2]
plt.figure(figsize=(15,15))
plt.plot(omega,n,label="Re(n)")
#plt.plot(omega,kappa,label="kappa")
plt.legend()
plt.xscale('log')
plt.savefig("dataPrJonas.png")
plt.show()
```
